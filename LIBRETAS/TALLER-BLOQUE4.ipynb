5. Filtrado condicional

````python
# Filtrar registros solo de México
df_mexico = df[df['countryCode'] == 'MX']
df_mexico.head()
````

````python
# Filtrar registros recientes (>= año 2000)
df_recientes = df[df['year'] >= 2000]
df_recientes.head()
````

````python
# Filtrar con condiciones múltiples: México y año reciente
df_mexico_2000 = df[(df['countryCode'] == 'MX') & (df['year'] >= 2000)]
df_mexico_2000.head()
````

7. Manejo de valores faltantes (NaN)

````python
# Contar valores faltantes
df.isna().sum()
````

````python
# Eliminar filas con valores faltantes en coordenadas
df_clean = df.dropna(subset=['depth', 'depthAccuracy'])
print("Registros originales:", len(df))
print("Registros con profundidad:", len(df_clean))
````
